```python
import requests
from bs4 import BeautifulSoup
import re
import numpy as np

Laptop_Brand = []
Model=[]
Display=[]
Storage=[]

Selling_Price=[]
Original_Price=[]
Discount=[]
Ratings=[]
Processor=[]
Ram=[]
for i in range(1,60): 
    url = f"https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}"
    requests_header = {'Content-Type': 'text/html; charset=UTF-8','User-Agent': 'Chrome/101.0.0.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/119.0','Accept-Encoding': 'gzip, deflate, br'}
    response=requests.get(url,headers=requests_header)
    soup = BeautifulSoup(response.text)  

    for item in soup.find_all('div', class_='yKfJKb'):  
        a = item.text
        
        brand= re.findall(r'^(.*?)\s', a)  
        if brand:
            Laptop_Brand.append(brand[0])  
        else:
            Laptop_Brand.append(np.nan) 
        m=re.findall(r'^(.*?)(?=\s-\s)',a)
        if m:
            Model.append(m[0])
        else:
            Model.append(np.nan)
        d=re.findall(r"((\d{1,2}\.\d{1,2}|\d{1,2})\s?(Inch|inch|INCH|cm|Cm|Cm))",a)
        if d:
            Display.append(d[1][0])
        else:
            Display.append(np.nan)
            
        s=re.findall(r"\/(\d+\s(?:TB|GB)\s?(?:SSD|HDD|EMMC\s?Storage)?)\/",a)    
        if s:
            Storage.append(s[0])
        else:
            Storage.append(np.nan)
        o=re.findall(r"₹([\d,]+)₹",a)
        
        if o:
            Selling_Price.append(o[0])
        else:
            Selling_Price.append(np.nan)
        p=re.findall(r"₹[\d,]+₹(\d{1,2}(?:,\d{2,3})*)",a)
        if p:
            Original_Price.append(p[0])
        else:
            Original_Price.append(np.nan)
        offer=re.findall(r"\d{1,2}%\s?off",a)
        if offer:
            Discount.append(offer[0])
        else:
            Discount.append(np.nan)
        r=re.findall(r"(\d(?:\.\d)?)[\d,]*\sRatings",a)
        if r:
            Ratings.append(r[0])
        else:
            Ratings.append(np.nan)
        processor=re.findall(r"Reviews(.*?)Processor",a)
        
        if  processor:
            Processor.append(processor[0])
        else:
            Processor.append(np.nan)
        ram=re.findall(r"((\d+ GB) (DDR4|DDR5|LPDDR4X|LPDDR5|LPDDR5X|Unified Memory)) RAM",a)
        if ram:
            Ram.append(ram[0][0])
        else:
            Ram.append(np.nan)

print(Laptop_Brand)
print(Model)
print(Display)
print(Storage)

print(Selling_Price)
print(Original_Price)
print(Discount)
print(Ratings)
print(Processor)
print(Ram)
```

```python
import pandas as pd
laptop=pd.DataFrame({"Brand":Laptop_Brand,
                   "Model":Model,
                   "Display":Display,
                   "Storage":Storage,
                   "Selling_Price":Selling_Price,
                   "Original_price":Original_Price,
                   "Discount":Discount,
                   "Ratings":Ratings,
                    "Processor":Processor,
                    "Ram":Ram})
laptop
```

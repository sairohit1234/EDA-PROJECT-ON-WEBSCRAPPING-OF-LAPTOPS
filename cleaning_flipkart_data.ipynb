```python

def convert_cm_to_inches(value):
    if 'cm' in value:
        cm_value = float(value.split(' ')[0])
        inches_value = cm_value * 0.393701
        return f"{round(inches_value, 2)} Inch"
    return value

df['Display'] = df['Display'].apply(convert_cm_to_inches)


df[['Display']].head()
```

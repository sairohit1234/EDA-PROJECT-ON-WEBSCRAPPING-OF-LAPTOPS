```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'Cleaned_laptop_data.csv'
data = pd.read_csv(file_path)
data
```

```python
data=data[data.Selling_Price<100000]
```

## Univariate Analysis

- Univariate Analysis on Categorical Columns

```python
data.Brand.value_counts()
```

```python
brand_counts = data['Brand'].value_counts()

plt.figure(figsize=(10, 6))
bars = plt.bar(brand_counts.index, brand_counts.values)

plt.title('Laptop Distribution by Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)


for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval),
             ha='center', va='bottom')  # ha: horizontal alignment, va: vertical alignment

plt.show()
```

**INSIGHTS**

The above code creates a bar chart to visualize the distribution of laptop brands in the dataset. Each brand is represented on the x-axis, and the corresponding count of laptops is shown on the y-axis. The exact count of each brand is displayed above the respective bars for clarity.

```python
brand_counts = data['Brand'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(brand_counts, labels=brand_counts.index, autopct='%1.1f%%',shadow=True)
plt.title('Laptop Distribution by Brand')
plt.axis('equal')
plt.show()
```

**INSIGHTS**

This code creates a pie chart to visualize the distribution of laptops by brand, with percentages displayed for each brand.The shadow is added for visual depth. The chart shows the relative market share of each brand in the dataset.

## Bivariate Analysis

```python

grouped_by_brand = data.groupby('Brand').agg({
    'Selling_Price': 'mean',
    'Original_price': 'mean',
    'discount_in_percentage': 'mean',
    'Ratings': 'mean',
    'display_in_inches': 'mean'
})


grouped_by_brand
```

```python

grouped_by_brand['Ratings'].plot(kind='line', figsize=(8, 4), title='Ratings')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python

grouped_by_brand['discount_in_percentage'].plot(kind='line', figsize=(8, 4), title='discount_in_percentage')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
grouped_by_brand['Selling_Price'].plot(kind='line', figsize=(8, 4), title='Selling_Price')
plt.gca().spines[['top', 'right']].set_visible(False)
```

The grouped analysis by Brand shows the following average values for continuous variables:

- Selling Price: Apple and Samsung have the highest average prices, while brands like CHUWI and Ultimus have the lowest.
- Discount in Percentage: ZEBRONICS offers the highest average discount, while Samsung has the lowest.
- Ratings: Apple and Colorful lead with higher ratings, while Ultimus has the lowest.
-Display Size: Most brands average around 15 inches, with Colorful and ZEBRONICS having 15.6 inches on average.

```python
pivot_table = pd.pivot_table(data, values=['Selling_Price', 'Ratings'],
                             index=['Ram'], aggfunc={'Selling_Price': 'mean', 'Ratings': 'mean'})


pivot_table
```

```python

pivot_table['Selling_Price'].plot(kind='line', figsize=(14, 4), title='Selling_Price')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python

pivot_table.plot(kind='scatter', x='Ratings', y='Selling_Price', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python

pivot_table['Ratings'].plot(kind='line', figsize=(14, 4), title='Ratings')
plt.gca().spines[['top', 'right']].set_visible(False)
```

- 32 GB DDR5 RAM models have the highest average selling price at ₹174,240 and also receive high ratings (4.65).
- 8 GB LPDDR5X models have a perfect rating of 5.0, but their average price is relatively lower at ₹32,990.
- 4 GB LPDDR4X models have the lowest average price (₹19,571.60) and relatively lower ratings.
